```python
import numpy as np
import matplotlib.pyplot as plt
```

Using the relationship $x_{n+1} = p \cdot (x_n) (1-x_n)$

Iterated through values of p from 0 to 4 in increments of 0.1. Performed 1000 iterations and plotted the next 100 iterations as +'s in the plot below.


```python
for p in np.arange(0, 4, 0.1):
  x = np.random.rand()
  for i in range(1000):
    x = p*x*(1-x)

  xs = []
  for i in range(300):
    x = p*x*(1-x)
    xs.append(x)

  plt.scatter(np.arange(len(xs)), xs,marker='+', s=1, label='p={:.1f}'.format(p))
plt.legend()
plt.show()
```

This above graph is a little difficult to see.

Plot a few cases first to try and get better understanding of the problem

```python
def generate_plot(p):
  
  x = np.random.rand()
  for i in range(1000):
    x = p*x*(1-x)

  xs = []
  for i in range(300):
    x = p*x*(1-x)
    xs.append(x)
  plt.scatter(np.arange(len(xs)), xs,marker='+', s=1, label='{:.2f}'.format(p))
  # plt.title('p={:.1f}'.format(p))
```

```python
for i in np.arange(0, 4, 0.1):
  generate_plot(i)
  plt.legend()
  plt.show()
```

Chaotic behaviour is observed for p higher than 3.6


```python
for i in np.arange(3.6, 4, 0.02):
  generate_plot(i)
  plt.legend()
  plt.show()
```

Interesting behaviour seems to occur between various ranges of p. For example, plotting 1000 points, gives concentrated points in some regions but other regions that are less concentrated
